### Introduction to Reinforcement Learning (RL)

In RL, we study the interaction between an **agent** and an **environment**. The agent takes actions to achieve a goal, guided by rewards from the environment. Our aim is to develop agents that can learn optimal behaviors through these interactions.



### Creating an Environment

An environment in RL defines the space in which the agent operates. It returns a new state and a reward for each action taken by the agent.

```python
class Env:
    def __init__(self):
        self.state = np.random.randint(2)
        self.terminated = False

    def step(self, action):
        if (action % 2 == self.state):
            reward = 1
        else:
            reward = -1
        self.state = np.random.randint(2)
        return self.state, reward, self.terminated, False, {}

    def reset(self):
        self.state = np.random.randint(2)
        self.terminated = False
        return self.state
```

### Building an Agent
Agents in RL decide which actions to take in an environment. A simple agent might act randomly or follow a predetermined policy.



```python
class Agent:
    def __init__(self, env):
        pass

    def act(self, state):
        return np.random.randint(2)
```

### Running an Experiment

To evaluate our agent's performance, we generate trajectories of state-action-reward sequences and compute the total reward.


```python
def run_experiment(env, agent, nb_steps):
    state = env.reset()
    res = [state]
    rewards = []
    for _ in range(nb_steps):
        action = agent.act(state)
        state, reward, terminated, truncated, info = env.step(action)
        res += [action, reward, state]
        rewards.append(reward)

    return res, rewards
```

```python
# --- Notebook-friendly rendering (pas de fenêtre X) ---
# Si tu veux quand même utiliser Xvfb, garde-le lancé, mais ce code n'en a pas besoin.

import time
from typing import Dict, Any, Optional

import numpy as np
import gymnasium as gym
from IPython.display import clear_output, display
from PIL import Image

# Si tu utilises Gym "legacy", installe gymnasium pour la compatibilité moderne.
# !pip install gymnasium[box2d]==0.29.1 ale-py==0.8.1 autorom
# !AutoROM --accept-license

def _fix_env_id(env_id: str) -> str:
    """Rend l'ID compatible gymnasium si nécessaire (ex: Atari)."""
    if env_id == "PongNoFrameskip-v4":
        # L'ID moderne côté Gymnasium/ALE
        return "ALE/Pong-v5"
    return env_id

def _pause_for_fps(start_t: float, fps: int):
    frame_time = 1.0 / max(fps, 1)
    dt = frame_time - (time.time() - start_t)
    if dt > 0:
        time.sleep(dt)

def exp_render_notebook(cfg: Dict[str, Any], policy: Optional[str] = "random"):
    """
    Affiche l'env dans le notebook:
    - cfg: {"name": str, "fps": int, "nb_step": int}
    - policy: "random" ou None (random par défaut)
    """
    env_id = _fix_env_id(cfg["name"])
    fps = int(cfg.get("fps", 20))
    nb_step = int(cfg.get("nb_step", 200))

    # Crée l'env en mode image pour rendu inline
    # NB: FrozenLake-v1 en rgb_array nécessite Gymnasium récent (sinon, enlever FrozenLake de la liste)
    try:
        env = gym.make(env_id, render_mode="rgb_array")
    except Exception as e:
        print(f"[SKIP] Impossible d'ouvrir {env_id} en rgb_array:\n{e}")
        return

    try:
        obs, info = env.reset()
        for t in range(nb_step):
            t0 = time.time()

            # Action (random policy par défaut)
            if policy == "random" or policy is None:
                action = env.action_space.sample()
            else:
                # place-holder si tu veux brancher ton agent
                action = env.action_space.sample()

            obs, reward, terminated, truncated, info = env.step(action)

            # Rendu image
            frame = env.render()  # np.ndarray (H, W, 3)
            if frame is None:
                print("[WARN] render() a renvoyé None")
                break

            # Affichage inline
            clear_output(wait=True)
            display(Image.fromarray(frame))

            if terminated or truncated:
                obs, info = env.reset()

            _pause_for_fps(t0, fps)
    finally:
        env.close()


# --- Ta config d'expériences ---
env_render_configs = [
    {"name": 'FrozenLake-v1',              "fps": 2,  "nb_step": 30},
    {"name": 'CartPole-v1',                "fps": 17, "nb_step": 120},
    {"name": 'LunarLanderContinuous-v3',   "fps": 30, "nb_step": 300},
    {"name": 'PongNoFrameskip-v4',         "fps": 40, "nb_step": 800},  # mappé -> ALE/Pong-v5
]

# --- Lancement séquentiel ---
for cfg in env_render_configs:
    print(f"▶️  {cfg['name']}  (fps={cfg['fps']}, steps={cfg['nb_step']})")
    exp_render_notebook(cfg)
```
